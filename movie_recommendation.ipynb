```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

## Pre-processing

```python
#Extract desire columns
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
import nltk
#to extract genres and keywords
def convert1(items):
    l = []
    for i in ast.literal_eval(items):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert1)
movies['keywords']=movies['keywords'].apply(convert1)
movies
```

```python
#to extract cast
def convert2(items):
    l = []
    count = 0
    for i in ast.literal_eval(items):
        if count != 3:
            l.append(i['name'])
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
movies
```

```python
#to extract director
def convert3(items):
    l = []
    for i in ast.literal_eval(items):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew']=movies['crew'].apply(convert3)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['overview']
```

```python
movies.head()
```

```python
movietags = movies[['movie_id','title','tags']]
```

```python
movietags['tags'] = movietags['tags'].apply(lambda x:" ".join(x))
```

```python
movietags['tags'] = movietags['tags'].apply(lambda x:x.lower())
```

```python
movietags.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#pip install nltk
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movietags['tags'] = movietags['tags'].apply(stem)
```

```python
movietags['tags'][1300]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movietags['tags']).toarray()
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0].shape
```

```python
def recommend(movie):
    movie_index = movietags[movietags['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movietags['title'][i[0]])
```

```python
recommend('Batman Begins')
```
